### Temperature Data Reformatting

```python
lithuania_temp_df = pd.read_csv("lithuania_1901_2016.csv")
russia_temp_df = pd.read_csv("russia_1901_2016.csv")
finland_temp_df = pd.read_csv("finland_1901_2016.csv")
france_temp_df = pd.read_csv("france_1901_2016.csv")
thailand_temp_df = pd.read_csv("thailand_1901_2016.csv")
south_africa_temp_df = pd.read_csv("south_africa_1901_2016.csv")
```

```python
# check the index of year 1984 (which is 1007 cuz each year has 12 data (12 months))
lithuania_temp_df.index[lithuania_temp_df[' Year'] == 1984].tolist()
```

```python
# check columns' names
lithuania_temp_df.columns
```

```python
# create a function for get the wanted data
def reformat_temp_data(dataframe):
    # rename the columns
    dataframe.columns = ['temperature(Celsius)', 'year', 'month', 'country', 'iso3']
    
    # keep the data from 1985 - 2016
    dataframe = dataframe[1008:]
    
    # drop the unused columns
    dataframe = dataframe.drop(['month', 'iso3'], axis=1)

    # get the average temperature each year per country
    dataframe = dataframe.groupby(['year', 'country']).mean()
    
    return dataframe
```

```python
lithuania_final_df = reformat_temp_data(lithuania_temp_df)
finland_final_df = reformat_temp_data(finland_temp_df)
russia_final_df = reformat_temp_data(russia_temp_df)
france_final_df = reformat_temp_data(france_temp_df)
thailand_final_df = reformat_temp_data(thailand_temp_df)
south_africa_final_df = reformat_temp_data(south_africa_temp_df)
```

```python
# concatenate
df = pd.concat([lithuania_final_df, finland_final_df, 
                russia_final_df, france_final_df, 
                thailand_final_df, south_africa_final_df])
df.to_csv('temp.csv')
```
